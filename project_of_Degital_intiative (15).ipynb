```python
#Assign a function
def hist1(x,y,c,h):
   import matplotlib.pyplot as plt
   import matplotlib.pyplot as plt
# Creating subplots with multiple histograms
   fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

   axes[0].hist(x, bins=30, color=c, edgecolor='black')
   axes[0].set_title('Histogram 1')

   axes[1].hist(y, bins=30, color=h, edgecolor='black')
   axes[1].set_title('Histogram 2')


# Adding labels and title
   for ax in axes:
	      ax.set_xlabel('range')
	      ax.set_ylabel('index')

# Adjusting layout for better spacing
   plt.tight_layout()

# Display the figure
   plt.show()

def histogram2(j,k,m,l):
    import matplotlib.pyplot as plt

# Creating a stacked histogram
    plt.hist([df[j],df[k]], bins=30, stacked=True, color=[m,l], edgecolor='black')

# Adding labels and title
    plt.xlabel('ranges')
    plt.ylabel('evaluation')
    plt.title('comparing')

# Adding legend
    plt.legend([j, k])

# Display the plot
    plt.show()


def hesto3(g,y,o):
		#Another histogram
    import matplotlib.pyplot as plt

# Plotting a basic histogram
    plt.hist(g, bins=30, color=o, edgecolor='black')

# Adding labels and title
    plt.xlabel('ranges')
    plt.ylabel('evaluation')
    plt.title(y)

# Display the plot
    plt.show()
```

# What range in the main players?

What range in player who wear 10?

```python
#Compare histograms in different ways
import matplotlib.pyplot as plt
import seaborn as sns
# Creating a customized histogram with a density plot
sns.histplot(df['away_player_10'], bins=30, kde=True, color='lightgreen', edgecolor='red')

# Adding labels and title
plt.xlabel('range')
plt.ylabel('evaluation')
plt.title('away_land')
# Display the plot
plt.show()

```

```python
#Compare histograms in different ways part2
sns.histplot(df['home_player_10'], bins=30, kde=True, color='brown', edgecolor='green')

# Adding labels and title
plt.xlabel('range')
plt.ylabel('evaluation')
plt.title('home_land')
```

conclusion:In the last matches, the visiting players were at the same level as the host players, unlike in the early matches, where the visiting players were better.
